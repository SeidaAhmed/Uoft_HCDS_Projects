Distribution Plot of log transformed continuous varioables, this type of visualization helps to see a given continuous variable follows normal distribution or not 
in this case I log transformed each continous varaible to make thir distribution normal so that we can able to implement linear regression for rental price prediction.

As the below distribution plot clearly depicts that the log tranformed continous variables in this case price and square feet have normaly distributed. We can also check for the rest of the variables using similar methods

```python

df["price"] = np.log1p(df["price"])
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
sns.distplot(df["price"], hist=True)
plt.legend(['Skewness={:.2f} Kurtosis={:.2f}'.format(
df["price"].skew(), 
df["price"].kurt())],
loc='best')
plt.title("Price")
plt.show()
```

```python
df["sqfeet"] = np.log1p(df["sqfeet"])
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
sns.distplot(df["sqfeet"], hist=True)
plt.legend(['Skewness={:.2f} Kurtosis={:.2f}'.format(
df["sqfeet"].skew(), 
df["sqfeet"].kurt())],
loc='best')
plt.title("Square Feet")
plt.show()
```

Linear Regression 

In order to predict the rental price I have implmented linear regression. I decided to apply this model based on two reasons. One the depndent variable which is price is continous variable and independent variables are also mostly numerical. Secondly, the dependent variable and other contnious variables are normaly distributed when log tranformed. As a result we can able to use linear regression.

```python
y=df['price']
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics


#separate the other attributes from the predicting attribute
x = df.drop('price',axis=1)
x=x.drop('region_url',axis=1)
x=x.drop('type',axis=1)



```

```python
from sklearn.preprocessing import LabelEncoder
# handle catagorical variable/ change it to numeric
number=LabelEncoder()
x["parking_options"]=number.fit_transform(x["parking_options"].astype("str"))
x["laundry_options"]=number.fit_transform(x["laundry_options"].astype("str"))
x["region"]=number.fit_transform(x["region"].astype("str"))
x["state"]=number.fit_transform(x["state"].astype("str"))

```

```python

# splitting the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

# creating an object of LinearRegression class
LR = LinearRegression()
# fitting the training data
LR.fit(x_train,y_train)
y_prediction =  LR.predict(x_test)
```

evaluate the model using mean abs. Based on the result the MAE is 32.2%

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_prediction)
```
